#MERGING DF IN ONE FINAL DF

```python
# 1. First, merge orderlines with orders
# We use 'id_order' from orderlines and 'order_id' from orders
df_merged = pd.merge(df_orderlines,
                     df_orders,
                     left_on='id_order',
                     right_on='order_id',
                     how='inner')

# 2. Then, merge the result with products using the 'sku' column
df_final = pd.merge(df_merged,
                    prodcuts_df_cleaned,
                    on='sku',
                    how='left') # 'left' ensures we keep orders even if a product is missing from df_products

# Optional: Clean up redundant columns
# Since you have 'id_order' and 'order_id' which are the same, you can drop one
df_final = df_final.drop(columns=['order_id'])

print(df_final.head())
```

```python
df_final.info()
```

# CALCULATING REVENUE AND DISCOUNT

```python
# 1. Revenue per line item (Price paid * quantity)
df_final['revenue'] = df_final['unit_price'] * df_final['product_quantity']

# 2. Price Difference (Absolute difference between list price and paid price)
df_final['price_diff'] = df_final['price'] - df_final['unit_price']

# 3. Discount Percentage
# Using .where or a simple calculation (handling potential division by zero)
df_final['discount_percent'] = (df_final['price_diff'] / df_final['price']) * 100

# 4. Additional Info: "Discount Category"
# This helps you group orders by how aggressive the sale was
def categorize_discount(pct):
    if pct <= 0: return 'Full Price'
    if pct <= 10: return 'Small Discount'
    if pct <= 30: return 'Standard Sale'
    return 'Clearance/Heavy Discount'

df_final['discount_type'] = df_final['discount_percent'].apply(categorize_discount)

# 5. Additional Info: Order Weight (Quantity per Order)
# Helps see if certain products are bought in bulk
df_final['is_bulk'] = df_final['product_quantity'] > 5

print(df_final[['sku', 'revenue', 'price_diff', 'discount_percent', 'discount_type']].head())
```
